```python
import pandas as pd
import numpy as np
```

### Quality of Life Dataset

```python
# This is the dataset with a variety of countries with the quality of life scores

df_quality = pd.read_csv('Quality_of_Life.csv')
df_quality.shape
```

```python
df_quality.info()
```

```python
df_quality.head()
```

```python
# Rename the country column

df_quality.rename(columns={"country": "Country"}, inplace = True)

df_quality
```

```python
# Drop unneeded columns
# We only need the country and Quality of Life Values from this dataset

df_quality.drop(columns = ["Purchasing Power Value", "Purchasing Power Category", "Safety Value", "Safety Category",
                           "Health Care Value", "Health Care Category", "Climate Value", "Climate Category",
                           "Cost of Living Value", "Cost of Living Category", "Property Price to Income Value",
                           "Property Price to Income Category", "Traffic Commute Time Value", "Traffic Commute Time Category",
                           "Pollution Value", "Pollution Category", "Quality of Life Category"], inplace = True)

df_quality
```

```python
# Selecting the rows without the 0.0

df_quality = df_quality[df_quality['Quality of Life Value'] != '0.0']

df_quality
```

```python
# Setting the index as Country

df_quality.set_index(["Country"], inplace = True)
df_quality.sort_index(inplace = True)

df_quality
```

```python
# Getting rid of the ': '

df_quality["Quality of Life Value"] = (df_quality["Quality of Life Value"]
                                       .astype(str) # This makes sure that it is a string for replacement
                                       .str.replace(":", "", regex=False) # This removes ":"
                                       .str.replace("'", "", regex=False) # This removes "'"
                                       .str.strip() # This removes any spaces
                                       .astype(float)) # This converts it to a float
df_quality
```

```python
# Checking if they are all converted to floats

df_quality.info()
```

```python
# Checking how many unique countries there are

df_quality.index.get_level_values("Country").unique()
```

### GDP Dataset

```python
# This is the GDP dataset

df_gdp = pd.read_csv('better-life-index-2024.csv')
df_gdp
```

```python
df_gdp.info()
```

```python
# Drop unneeded columns
# We only need the country and GDP per capita (USD) from this dataset

df_gdp.drop(columns = ["  Dwellings without basic facilities", "  Housing expenditure", "  Rooms per person",
                       "  Household net adjusted disposable income", "  Household net wealth", "  Labour market insecurity",
                       "  Employment rate", "  Long-term unemployment rate", "  Personal earnings", "  Quality of support network",
                       "  Educational attainment", "  Student skills", "  Years in education", "  Air pollution",
                       "  Water quality", "  Stakeholder engagement for developing regulations", "  Voter turnout",
                       "  Life expectancy", "  Self-reported health", "  Feeling safe walking alone at night", "  Homicide rate",
                       "  Employees working very long hours", "  Time devoted to leisure and personal care", "  Life satisfaction"], inplace = True)

df_gdp
```

```python
# Rename GDP per capita (USD)

df_gdp.rename(columns={"GDP per capita (USD)": "GDP per Capita (USD)"}, inplace = True)

df_gdp
```

```python
# Setting the index as Country

df_gdp.set_index(["Country"], inplace = True)
df_gdp.sort_index(inplace = True)

df_gdp
```

```python
df_gdp.info()
```

### Happiness Dataset

```python
# This is the Happiness dataset

df_happiness = pd.read_csv('2019.csv')
df_happiness
```

```python
# Rename Country or region as Country

df_happiness.rename(columns={"Country or region": "Country", "Score": "Happiness Score"}, inplace = True)
df_happiness
```

```python
df_happiness.info()
```

```python
# Dropping columns other than Country and Happiness Score

df_happiness.drop(columns = ["Overall rank", "GDP per capita", "Social support", "Healthy life expectancy",
                             "Freedom to make life choices", "Generosity", "Perceptions of corruption"], inplace = True)

df_happiness
```

```python
# Setting the index as Country

df_happiness.set_index(["Country"], inplace = True)
df_happiness.sort_index(inplace = True)

df_happiness
```

```python
# Merge the Quality of Life dataset and the GDP dataset first

merged_df = df_quality.merge(df_gdp, on = "Country", how = "inner")

# Then, merge with the Happiness dataset

merged_df = merged_df.merge(df_happiness, on = "Country", how = "inner")

merged_df
```

```python
merged_df.info()
```

### Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# A heatmap to compare the GDP, Quality of Life, and Happiness for the Countries

plt.figure(figsize=(12, 8))
sns.heatmap(pd.DataFrame(merged_df).sort_values(by="GDP per Capita (USD)", ascending = False), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout(pad=2)
plt.title('Impact of GDP per Capita on Quality of Life and Happiness Across Countries', fontsize=16, y=1.03)
plt.ylabel('Country', fontsize=12, labelpad=15)
plt.xlabel('Metrics', fontsize=12, labelpad=10)
plt.show()
```

```python
# A scatterplot to show the relationship between GDP and Quality of Life

plt.figure(figsize=(8,6))
plt.scatter(merged_df["Quality of Life Value"],merged_df["GDP per Capita (USD)"])
plt.xlabel("Quality of Life Value")
plt.ylabel("GDP per Capita (USD)")
plt.title("Correlation Between GDP per Capita and Quality of Life")
```

```python
# A scatterplot to show the relationship between GDP and Happiness

plt.figure(figsize=(8,6))
plt.scatter(merged_df["Happiness Score"],merged_df["GDP per Capita (USD)"])
plt.xlabel("Happiness Score")
plt.ylabel("GDP per Capita (USD)")
plt.title("Correlation Between GDP per Capita and Happiness Score")
```

```python
# A merged scatterplot of GDP vs Quality of Life and Happiness Score

import matplotlib.pyplot as plt

fig, primary_x_axis = plt.subplots(figsize=(10, 6))

# Quality of Life scatterplot
quality_scatter = primary_x_axis.scatter(merged_df["Quality of Life Value"],merged_df["GDP per Capita (USD)"],color='blue',alpha=0.8,label='Quality of Life',marker='o')

primary_x_axis.set_xlabel('Quality of Life Value', color='blue', fontsize=12)
primary_x_axis.set_ylabel('GDP per Capita (USD)', fontsize=12)
primary_x_axis.tick_params(axis='x', labelcolor='blue')

# Happiness scatterplot
secondary_x_axis = primary_x_axis.twiny()

happiness_scatter = secondary_x_axis.scatter(merged_df["Happiness Score"],merged_df["GDP per Capita (USD)"],color='red',alpha=0.8,label='Happiness Score',marker='o')

secondary_x_axis.set_xlabel('Happiness Score', color='red', fontsize=12)
secondary_x_axis.tick_params(axis='x', labelcolor='red')

# x-axis limits so it is aligned into the same graph
primary_x_axis.set_xlim(merged_df["Quality of Life Value"].min() - 5, merged_df["Quality of Life Value"].max() + 5)

secondary_x_axis.set_xlim(merged_df["Happiness Score"].min() - 0.3, merged_df["Happiness Score"].max() + 0.3)


plt.title('GDP per Capita vs Quality of Life & Happiness Score', pad=20, fontsize=14)
primary_x_axis.grid(True, linestyle='--', alpha=0.4)

plt.legend([quality_scatter, happiness_scatter],['Quality of Life', 'Happiness Score'],loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# A scatterplot to compare GDP and happiness, with Quality of life being the size of the points

import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.patches as mpatches

# Normalize bubble size for a better scale
quality_scaled = merged_df["Quality of Life Value"] / merged_df["Quality of Life Value"].max() * 300

# s is the bubble size
plt.figure(figsize=(10, 6))
plt.scatter(merged_df["Happiness Score"], merged_df["GDP per Capita (USD)"], s=quality_scaled, color='blue', alpha=0.6, edgecolors='black', linewidth=0.5)

plt.xlabel("Happiness Score", fontsize=12)
plt.ylabel("GDP per Capita (USD)", fontsize=12)
plt.title("GDP per Capita vs Happiness Score (Bubble Size = Quality of Life)", fontsize=14, pad=15)
plt.grid(True, linestyle='--', alpha=0.3)

# Legend bubbles
for value in [100, 400, 700]:
    plt.scatter([], [], s=value / merged_df["Quality of Life Value"].max() * 300, color='blue', alpha=0.6, edgecolors='black', label=f'Quality of Life: {value}')

plt.legend(title="Quality of Life", loc='upper left', frameon=True)

plt.tight_layout()
plt.show()
```

```python
# A scatterplot comparing GDP and happiness, with quality of life being the quality of life values

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
scatter = plt.scatter(merged_df["Happiness Score"],merged_df["GDP per Capita (USD)"],c=merged_df["Quality of Life Value"],cmap='turbo',alpha=0.7,edgecolors='black',linewidth=0.5)

# Color bar
cbar = plt.colorbar(scatter)
cbar.set_label("Quality of Life Value")

plt.xlabel("Happiness Score", fontsize=12)
plt.ylabel("GDP per Capita (USD)", fontsize=12)
plt.title("GDP vs Happiness Score (Color = Quality of Life)", fontsize=14, pad=15)
plt.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# A heatmap to see how the metrics correlate with each other

plt.figure(figsize=(10,6))

sns.heatmap(merged_df[["GDP per Capita (USD)", "Quality of Life Value", "Happiness Score"]].corr(), annot=True, cmap='coolwarm')
```

```python
# A bar chart for the top 10 countries to show how GDP affects the happiness, using colors

import matplotlib.pyplot as plt
import seaborn as sns

# Select top 10 countries by GDP 
top10_gdp = merged_df.sort_values("GDP per Capita (USD)", ascending=False).head(10).reset_index()

plt.figure(figsize=(14, 6))
barplot = sns.barplot(data=top10_gdp, y="Country", x="GDP per Capita (USD)", hue="Happiness Score", dodge=False, palette="viridis")

plt.title("Top 10 Countries by GDP per Capita\nColored by Happiness Score", fontsize=16)
plt.xlabel("GDP per Capita (USD)", fontsize=12)
plt.ylabel("Country", fontsize=12)

for container in barplot.containers:
    barplot.bar_label(container, fmt='%.0f', label_type='edge', padding=3)

plt.tight_layout()
plt.show()
```

```python
# A bar chart for the top 10 countries to show how GDP affects the quality of life, using colors

import matplotlib.pyplot as plt
import seaborn as sns

# Sort and select top 10 countries by GDP 
top10_gdp = merged_df.sort_values("GDP per Capita (USD)", ascending=False).head(10).reset_index()

plt.figure(figsize=(16, 7))
barplot = sns.barplot(data=top10_gdp,y="Country",x="GDP per Capita (USD)",hue="Quality of Life Value", dodge=False, palette="viridis")

plt.title("Top 10 Countries by GDP per Capita\nColored by Quality of life Value", fontsize=16)
plt.xlabel("GDP per Capita (USD)", fontsize=12)
plt.ylabel("Country", fontsize=12)

for container in barplot.containers:
    barplot.bar_label(container, fmt='%.0f', label_type='edge', padding=3)

plt.tight_layout()
plt.show()
```

```python
# A quadrant analysis of how GDP affects the happiness score for each country

merged_df = merged_df.reset_index()
merged_df["Country"]

import matplotlib.pyplot as plt

x = merged_df["GDP per Capita (USD)"]
y = merged_df["Happiness Score"]
countries = merged_df["Country"]

# Calculate median for quadrants
x_median = x.median()
y_median = y.median()

plt.figure(figsize=(14, 9))
scatter = plt.scatter(x, y, alpha=0.7, color="cornflowerblue", edgecolors="black")

# Quadrant lines
plt.axhline(y=y_median, color='gray', linestyle='--', label=f'Median Happiness: {y_median:.2f}')
plt.axvline(x=x_median, color='gray', linestyle='--', label=f'Median GDP: {x_median:,.0f}')

# Labeling every country slightly above the point
for i in range(len(countries)):
    plt.text(x.iloc[i], y.iloc[i] + 0.03, countries.iloc[i], fontsize=8, ha='center')

plt.xlabel("GDP per Capita (USD)", fontsize=12)
plt.ylabel("Happiness Score", fontsize=12)
plt.title("Quadrant Analysis: GDP vs Happiness Score", fontsize=15, pad=10)
plt.grid(True, linestyle='--', alpha=0.3)

plt.legend()
plt.tight_layout()
plt.show()

# Q1: High GDP and High Happiness
# Q2: Low GDP and High Happiness
# Q3: Low GDP and Low Happiness
# Q4: High GDP and Low Happiness
```

```python
# A quadrant analysis of how GDP affects the quality of life for each country

import matplotlib.pyplot as plt

x = merged_df["GDP per Capita (USD)"]
y = merged_df["Quality of Life Value"]
countries = merged_df["Country"]

# Calculate medians
x_median = x.median()
y_median = y.median()

plt.figure(figsize=(14, 9))
plt.scatter(x, y, color='teal', edgecolors='black', alpha=0.7)

# Quadrant lines
plt.axhline(y=y_median, color='gray', linestyle='--', label=f'Median Quality of Life: {y_median:.1f}')
plt.axvline(x=x_median, color='gray', linestyle='--', label=f'Median GDP: {x_median:,.0f}')

# Label each country
for i in range(len(countries)):
    plt.text(x.iloc[i], y.iloc[i] + 1.2, countries.iloc[i], fontsize=8, ha='center')

plt.xlabel("GDP per Capita (USD)", fontsize=12)
plt.ylabel("Quality of Life Value", fontsize=12)
plt.title("Quadrant Analysis: GDP per Capita vs Quality of Life", fontsize=14, pad=15)
plt.grid(True, linestyle='--', alpha=0.3)

plt.legend()
plt.tight_layout()
plt.show()

# Q1: High GDP and High Life Quality
# Q2: Low GDP and High Life Quality
# Q3: Low GDP and Low Life Quality
# Q4: High GDP and Low Life Quality
```

```python
# A merged quadrant analysis of the 2 above

import matplotlib.pyplot as plt

gdp = merged_df["GDP per Capita (USD)"]
qol = merged_df["Quality of Life Value"]
happiness = merged_df["Happiness Score"]
countries = merged_df["Country"]

fig, ax1 = plt.subplots(figsize=(14, 10))

# Quality of Life on left y-axis
qol_plot = ax1.scatter(gdp, qol, color='seagreen', edgecolors='black', alpha=0.7, label="Quality of Life")
ax1.set_ylabel("Quality of Life Value", color='seagreen', fontsize=12)
ax1.tick_params(axis='y', labelcolor='seagreen')

# Quality of Life points
for i in range(len(countries)):
    ax1.text(gdp.iloc[i], qol.iloc[i] + 1.5, countries.iloc[i], fontsize=8, color='darkgreen', ha='center')

# Median GDP line
gdp_median = gdp.median()
ax1.axvline(x=gdp_median, color='gray', linestyle='--', label=f'Median GDP: {gdp_median:,.0f}')

# Happiness on right y-axis
ax2 = ax1.twinx()
happy_plot = ax2.scatter(gdp, happiness, color='navy', edgecolors='black', alpha=0.7, label="Happiness Score", marker='s')
ax2.set_ylabel("Happiness Score", color='navy', fontsize=12)
ax2.tick_params(axis='y', labelcolor='navy')

# Happiness points
for i in range(len(countries)):
    ax2.text(gdp.iloc[i], happiness.iloc[i] - 0.08, countries.iloc[i], fontsize=8, color='navy', ha='center')

ax1.set_xlabel("GDP per Capita (USD)", fontsize=12)
plt.title("GDP per Capita vs Happiness Score and Quality of Life (Dual Y-Axis)", fontsize=15, pad=15)
ax1.grid(True, linestyle='--', alpha=0.3)

handles = [qol_plot, happy_plot]
labels = ["Quality of Life", "Happiness Score"]
ax1.legend(handles, labels, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# A bar chart using plotly to compare GDP, happiness, and quality of life for the top 10 GDP countries
# Normalized so the 3 metrics can be on the same graph

import pandas as pd
import plotly.express as px
from sklearn.preprocessing import MinMaxScaler

# Top 10 countries by GDP
top10_gdp = merged_df.nlargest(10, 'GDP per Capita (USD)').copy()

# Normalizing the values
scaler = MinMaxScaler()
top10_gdp[['GDP per Capita (USD)', 'Happiness Score', 'Quality of Life Value']] = scaler.fit_transform(top10_gdp[['GDP per Capita (USD)', 'Happiness Score', 'Quality of Life Value']])

# Melting the DataFrame into long format
normalized_grouped_df = pd.melt(top10_gdp, id_vars='Country', value_vars=['GDP per Capita (USD)', 'Happiness Score', 'Quality of Life Value'],var_name='Metric', value_name='Normalized Value')

# Using Plotly to create grouped bar chart
fig = px.bar(normalized_grouped_df, x='Country', y='Normalized Value', color='Metric', barmode='group',title='Normalized Comparison of GDP, Happiness Score, and Quality of Life (Top 10 GDP Countries)',
             labels={'Normalized Value': 'Normalized Value (0 to 1)'}, height=500)

fig.show()
```
